```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
full_grouped_last = full_grouped_df[full_grouped_df["Date"] == full_grouped_df["Date"].max()]
```

```python
full_grouped_last.head()
```

```python
full_grouped_last.info()
```

```python
full_grouped_last["death_rate"] = full_grouped_last["Deaths"] *100 / full_grouped_last["Confirmed"]
```

```python
full_grouped_last["Recovery_rate"] = full_grouped_last["Recovered"] * 100/ full_grouped_last["Confirmed"]   
```

```python
full_grouped_last.head()
```

```python
full_grouped_last.describe()
```

```python
full_grouped_last["High_mortality"] = (full_grouped_last["death_rate"] > 2.1).astype("int") # from the statistics above 50% death rate is at 2.151, so I tagged high mortality above 2.1
```

```python
x = full_grouped_last[["Confirmed", "Recovered", "Deaths", "Active"]]
y = full_grouped_last["High_mortality"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 42)
```

```python
model = RandomForestClassifier(random_state = 42)
model.fit(x_train, y_train)
```

```python
preditions = model.predict(x_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, preditions))
print(classification_report(y_test, preditions))
```
